## Вирішення транспортної задачі за допомогою Python і SciPy

### Вступ

Транспортна задача — це спеціальний випадок лінійного програмування, де необхідно мінімізувати витрати на транспортування товарів з декількох пунктів відправлення (наприклад, заводів) до декількох пунктів призначення (наприклад, складів), враховуючи обмеження на обсяги поставок та потреб.

### Постановка задачі

Нехай у нас є 2 заводи, які виробляють однаковий продукт, і 3 склади, куди цей продукт потрібно доставити. Витрати на транспортування одиниці продукту від кожного заводу до кожного складу відомі, а також відомі щоденні виробничі потужності заводів і щоденні потреби складів.

**Витрати на транспортування (одиниця продукту):**

| Завод\Склад | Склад 1 | Склад 2 | Склад 3 |
|-------------|---------|---------|---------|
| Завод 1     | 2       | 3       | 1       |
| Завод 2     | 5       | 4       | 8       |

**Виробничі потужності:**

- Завод 1: 50 одиниць
- Завод 2: 60 одиниць

**Потреби складів:**

- Склад 1: 20 одиниць
- Склад 2: 30 одиниць
- Склад 3: 60 одиниць

### Рішення задачі за допомогою SciPy

```python
import numpy as np
from scipy.optimize import linprog

# Витрати на транспортування
cost = np.array([[2, 3, 1], [5, 4, 8]])

# Виробничі потужності і потреби складів
supply = [50, 60]
demand = [20, 30, 60]

# Перетворення задачі в стандартну форму для linprog
c = cost.flatten()
A_eq = []
for i in range(len(supply)):
    row = [0]*len(c)
    for j in range(len(demand)):
        row[i*len(demand)+j] = 1
    A_eq.append(row)

for j in range(len(demand)):
    row = [0]*len(c)
    for i in range(len(supply)):
        row[i*len(demand)+j] = 1
    A_eq.append(row)

b_eq = supply + demand

# Розв'язок задачі
res = linprog(c, A_eq=A_eq, b_eq=b_eq, method='highs')

print("Вартість транспортування:", res.fun)
print("Оптимальний план по

ставок:", res.x.reshape((len(supply), len(demand))))
```

### Аналіз результатів

Після виконання коду ми отримаємо оптимальний план поставок, який мінімізує загальні витрати на транспортування, а також загальну вартість. Результат допоможе нам розуміти, як ефективно розподіляти ресурси між виробничими потужностями та споживачами.

---

Цей приклад демонструє, як можна використовувати Jupyter Notebook для навчання студентів рішенню реальних задач за допомогою програмування та математичного моделювання. Такий підхід сприяє кращому засвоєнню матеріалу та розвитку практичних навичок.

````python
import numpy as np
from scipy.optimize import linprog

# Витрати на транспортування одиниці продукту від кожного заводу до кожного складу
# Матриця витрат має розмірність 2x3, де рядки представляють заводи, а стовпці - склади
cost = np.array([
    [2, 3, 1],  # Витрати заводу 1 до складів 1, 2, 3 відповідно
    [5, 4, 8]   # Витрати заводу 2 до складів 1, 2, 3 відповідно
])

# Виробничі потужності заводів (кількість одиниць, які кожен завод може поставити)
supply = [50, 60]  # Завод 1 може поставити 50 од., Завод 2 може поставити 60 од.

# Потреби складів (кількість одиниць, необхідних кожному складу)
demand = [20, 30, 60]  # Складам 1, 2, 3 потрібно 20, 30, 60 од. відповідно

# Лінійне програмування вимагає роботи з вектором вартостей, тому перетворюємо матрицю витрат в одновимірний масив
c = cost.flatten()

# Побудова матриці обмежень. Для кожного заводу і складу створюється рядок в матриці обмежень,
# що відображає обмеження на поставку або потребу.
A_eq = []

# Додавання обмежень за виробничими потужностями
for i in range(len(supply)):
    row = [0]*len(c)
    for j in range(len(demand)):
        row[i*len(demand)+j] = 1
    A_eq.append(row)

# Додавання обмежень за потребами складів
for j in range(len(demand)):
    row = [0]*len(c)
    for i in range(len(supply)):
        row[i*len(demand)+j] = 1
    A_eq.append(row)

# Об'єднання обмежень за виробничими потужностями та потребами складів в один список
b_eq = supply + demand

# Виконання оптимізації з використанням linprog. Ця функція знайде мінімальну вартість, 
# виходячи з витрат на транспортування, обмежень поставок та потреб.
res = linprog(c, A_eq=A_eq, b_eq=b_eq, method='highs')

# Виведення результатів
print("Вартість транспортування:", res.fun)  # Загальна вартість транспортування
print("Оптимальний план поставок:\n", res.x.reshape((len(supply), len(demand))))  # Оптимальні кількості продукції для транспортування
````

Для виведення результатів оптимізації у формі таблиці та стовпчастої діаграми, ми можемо використовувати бібліотеку Pandas для створення та відображення таблиці, а Matplotlib для генерації діаграми. Нижче наведено код, який робить це, виходячи з результатів оптимізації, отриманих за допомогою `linprog`.




````python
import pandas as pd
import matplotlib.pyplot as plt

# Перетворення результатів оптимізації в матрицю для зручності подальшої роботи
optimal_plan_matrix = res.x.reshape((len(supply), len(demand)))

# Створення DataFrame для кращого відображення даних
df = pd.DataFrame(optimal_plan_matrix, 
                  index=[f'Завод {i+1}' for i in range(len(supply))], 
                  columns=[f'Склад {i+1}' for i in range(len(demand))])

print("Оптимальний план поставок:")
print(df)

# Генерація стовпчастої діаграми для візуалізації оптимального плану поставок
df.plot(kind='bar', figsize=(10, 6), alpha=0.75, rot=0)
plt.title('Оптимальний план поставок від заводів до складів')
plt.xlabel('Завод')
plt.ylabel('Кількість продукції')
plt.xticks(range(len(supply)), [f'Завод {i+1}' for i in range(len(supply))])
plt.legend(title='Склад')
plt.show()

````



Цей код спочатку перетворює оптимальний план поставок, отриманий з `linprog`, в матрицю, що відображає кількість продукції, яка повинна бути доставлена від кожного заводу до кожного складу. Далі, використовуючи бібліотеку Pandas, ми створюємо `DataFrame`, який зручно відображає ці дані у формі таблиці. Нарешті, за допомогою Matplotlib ми генеруємо стовпчасту діаграму, яка візуалізує кількість продукції, що повинна бути поставлена від кожного заводу до кожного складу, дозволяючи легко оцінити обсяги поставок візуально.
